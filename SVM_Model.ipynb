```python
import numpy as np
```

SVM classifier

```python
class SVM_classifier():
    def __init__(self , no_of_iterations , learning_rate , lambda_parameter):
        self.no_of_iterations = no_of_iterations
        self.learning_rate = learning_rate
        self.lambda_parameter = lambda_parameter
    
    def fit(self , X , Y):
        self.m , self.n = X.shape

        self.w = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.Y = Y

        for i in range(self.no_of_iterations):
            self.update_weights()

    def update_weights(self):

        #Label Encoding since in SVM we classify classes as 1's and -1's. But in general dataset have 1's and 0's. 
        #If not then skip the below step of y_label

        y_label = np.where(self.Y == 0 , -1 , 1)

        for index , x_i in range(self.X):
            condition = y_label[index] * (np.dot(self.w * x_i) - self.b)

            if(condition == True):
                #dw = 2*lambda*w
                #db = 0
                dw = 2 * self.learning_rate * self.w
                db = 0
            
            else:
                #dw = 2*lambda*w - (yi . xi)
                #db = yi
                dw = 2 * self.learning_rate * self.w - np.dot(y_label[index] , x_i)
                db = y_label[index]
            
            self.w = self.w - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db

    def predict(self , X):
        output = np.dot(self.w , X) - self.b
        predicted_label = np.sign(output)
        y_hat = np.where(predicted_label <= -1 , 0 , 1)
        return y_hat
```

```python
model = SVM_classifier(no_of_iterations = 1000 , learning_rate = 0.001 , lambda_parameter = 0.01)
```
